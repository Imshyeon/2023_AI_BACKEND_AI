# I. Data Preprocessing

```
import warnings
warnings.filterwarnings('ignore')
```

> ## 1) 'notExercise.xlsx' 파일 읽어오기

```
import pandas as pd

url = 'https://raw.githubusercontent.com/rusita-ai/pyData/master/notExercise.xlsx'
DF = pd.read_excel(url)

DF.head()
```

* DataFrame 정보

```
DF.info()
```

> ## 2) '기간' 열(Column) 삭제

* '기간'열(Column) 정보

```
DF['기간'].value_counts()
```

* '기간'열(Column) <font color = 'red'>삭제

```
del DF['기간']
```

* <font color = 'red'>삭제</font> 결과 확인

```
DF.head()
```

> ## 3) 0, 23~52번(지역분류) index 삭제

* '대분류' 정보

```
DF['대분류']
```

* 23번 ~ 52번 index <font color = 'red'>삭제

```
for i in [range(23,53)]:
  type(i)
  DF.drop(i, inplace=True)
```

```
DF
```

* 0번 index <font color = 'red'>삭제

```
for i in [range(0,1)]:
  print(type(i))
  DF.drop(i, inplace=True)
```

```
DF
```

# II. '성별' 통계

> ## 1) '성별' 데이터 추출

```
DF1=DF.iloc[0:2,0:7]
```

```
DF1
```

> ## 2) '대분류' 열(Column) 삭제

```
del DF1['대분류']
```

```
DF1
```

> ## 3) '분류'열(Column)을 index로 지정

```
DF1.set_index('분류',inplace = True)
```

```
DF1
```

> ## 4) matplotlib 시각화

```
import matplotlib.pyplot as plt
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (15, 10))
ax[0, 0].pie(DF1['운동을 할 충분한 시간이 없어서'].value_counts().index,
             labels = DF1['운동을 할 충분한 시간이 없어서'].index,
             autopct = '%.1f%%', startangle = 90,)

ax[0, 1].pie(DF1['함께 운동을 할 사람이 없어서'].value_counts().index,
             labels = DF1['함께 운동을 할 사람이 없어서'].index,
             autopct = '%.1f%%', startangle = 90,)

ax[1, 0].pie(DF1['운동을 할 만한 장소가 없어서'].value_counts().index,
             labels = DF1['운동을 할 만한 장소가 없어서'].index,
             autopct = '%.1f%%', startangle = 90,)

ax[1, 1].pie(DF1['운동을 싫어해서'].value_counts().index,
             labels = DF1['운동을 싫어해서'].index,
             autopct = '%.1f%%', startangle = 90,)

ax[0, 0].set_title('운동을 할 충분한 시간이 없어서')
ax[0, 1].set_title('함께 운동을 할 사람이 없어서')
ax[1, 0].set_title('운동을 할 만한 장소가 없어서')
ax[1, 1].set_title('운동을 싫어해서')

plt.show()
```

# III. '연령별' 통계

> ## 1) '연령별' 데이터 전처리

```
DF2=DF.iloc[2:8,0:7]
del DF2['대분류']
DF2.set_index('분류',inplace=True)
```

```
DF2
```

> ## 2) matplotlib 시각화

```
# import matplotlib.pyplot as plt
fig, ax = plt.subplots(nrows = 1, ncols = 3, figsize = (20,5))

print(DF2.columns[2])

values_1=DF2['운동을 할 충분한 시간이 없어서'].value_counts().index
labels_1=DF2['운동을 할 충분한 시간이 없어서'].index
ax[0].pie(values_1,labels = labels_1,autopct = '%.1f%%', startangle = 90,)

values_2=DF2['함께 운동을 할 사람이 없어서'].value_counts().index
labels_2=DF2['함께 운동을 할 사람이 없어서'].index
ax[1].pie(values_2, labels = labels_2, autopct = '%.1f%%', startangle = 90,)

values_3=DF2['운동을 할 만한 장소가 없어서'].value_counts().index
labels_3=DF2['운동을 할 만한 장소가 없어서'].index
ax[2].pie(values_3, labels = labels_3, autopct = '%.1f%%', startangle = 90,)


ax[0].set_title('운동을 할 충분한 시간이 없어서')
ax[1].set_title('함께 운동을 할 사람이 없어서')
ax[2].set_title('운동을 할 만한 장소가 없어서')

plt.show()
```
