```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import seaborn as sns
DF=sns.load_dataset('mpg')
```

```python
DF.info()
```

```python
DF.head()
```

```python
X=DF[['weight']]
y=DF['mpg']
```

```python
X[:5]
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(9,6))
sns.regplot(x=X, y=y)
plt.xlabel('weight_Without_Scailing')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler1=MinMaxScaler()
x_Norm = scaler1.fit_transform(X)
```

```python
x_Norm[:5]
```

```python
fig = plt.figure(figsize=(9,6))
sns.regplot(x=x_Norm, y=y)
plt.xlabel('weight_With_Normalization')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler2 = StandardScaler()
X_Stan = scaler2.fit_transform(X)
```

```python
X_Stan[:5]
```

```python
fig=plt.figure(figsize=(9,6))
sns.regplot(x=X_Stan, y=y)
plt.xlabel('weight_With_Standardization')
plt.show()
```
